```python
ProdName=[]
Discount=[]
Price=[]
Offer=[]
Features=[]
Ratings=[]
pageno=[]
```

```python
for i in range(1,28):
    start_time=time.time()
    url='https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&otracker=categorytree&sort=popularity'
    page=requests.get(url)
    pagecontent=page.text
    up=BeautifulSoup(pagecontent)
    
    for x in soup.find_all('div',attrs={'class':'_3pLy-c row'}):
        pname=x.find('div',attrs={'class':'_4rR01T'})
        cost=x.find('div',attrs={'class':'_30jeq3'})
        rat=x.find('div',attrs={'class':'_3LWZlK'})
        specs=x.find('ul',attrs={'class':'_1xgFaf'})
        Dis=x.find('div',attrs={'class':'_3Ay6Sb'})
        off=x.find('div',attrs={'class':'_27UcVY'})
    
        if pname is None:
            ProdName.append(np.NaN)
        else:
            ProdName.append(pname.text)
        if cost is None:
            Price.append(np.NaN)
        else:
            Price.append(cost.text)
        if rat is None:
            Ratings.append(np.NaN)
        else:
            Ratings.append(rat.text)
        if specs is None:
            Features.append(np.NaN)
        else:
            Features.append(specs.text)
        if Dis is None:
            Discount.append(np.NaN)
        else:
            Discount.append(Dis.text) 
        if off is None:
            Offer.append(np.NaN)
        else:
            Offer.append(off.text)    
        pageno.append(i)
    print('page {} completed in {} seconds'.format(i,time.time()-start_time))    
```

```python
MobiData=pd.DataFrame({'Product Name':ProdName,'OfferPrice':Price,'OrignalPrice':Offer,'Discount':Discount,'Ratings':Ratings,'Features':Features,'Page':pageno})
```
